```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from numpy.polynomial.polynomial import polyfit
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('C:\\Users\\deepa\\Downloads\\delivery_time.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
x = df['Sorting Time']
y = df['Delivery Time']
```

```python
plt.figure(figsize=(6,4))
b, m = polyfit(x, y, 1)
plt.plot(x, y, '.')
plt.plot(x, b + m * x, '-')
plt.scatter(x,y)
plt.xlabel("Sorting Time")
plt.ylabel("Delivery Time")
plt.show()
```

```python
corr = np.corrcoef(x, y)
```

```python
#Regression Model ,1 - No transformation
model = sm.OLS(y, x).fit()
predictions = model.predict(x)
```

```python
model.summary()
```

```python
#Log Transformation of X (model-2)
x_log = np.log(df['Sorting Time'])
model = sm.OLS(y, x_log).fit()
predictions = model.predict(x_log)
model.summary()
```

```python
#Log Transformation of Y (model-3)
y_log = np.log(df['Delivery Time'])
model = sm.OLS(y_log, x).fit()
predictions = model.predict(x)
model.summary()
```

```python
#Log Transformation of X & Y - model 4
model = sm.OLS(y_log, x_log).fit()
predictions = model.predict(x_log)
model.summary()
```

```python
#Sq Root Transformation of X (model-5)
x_sqrt = np.sqrt(df['Sorting Time'])
model = sm.OLS(y, x_sqrt).fit()
predictions = model.predict(x_sqrt)
model.summary()
```

```python
#Square Root Transformation of Y (model-6)
y_sqrt = np.sqrt(df['Delivery Time'])
model = sm.OLS(y_sqrt, x).fit()
predictions = model.predict(x)
model.summary()
```

```python
#Square Root Transformation of X & Y (model-7)
model = sm.OLS(y_sqrt, x_sqrt).fit()
predictions = model.predict(x_sqrt)
model.summary()
```

```python
#We will use Model 7 as it has the best R square value

# p-value < 0.01 :Thus the model is accepted

# coefficient == 1.64: Thus if the value of Sorting Time is increased by 1, the predicted value of Delivery Time will increase by 1.64

# Adj. R-sqared == 0.987 :Thus the model explains 98.7% of the variance in dependent variable
```
